```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import *
import random
```

```python
muscat = 'Muscat'
sitamarhi = 'Sitamarhi'
```

```python
def plot_yes_no(data, colors) :
    labels = ['Yes','No']
    plt.tight_layout()
    plt.pie(data, labels = labels, colors = colors, autopct = lambda p: '{:.2f}%\n({:.0f})'.format(p, (p/100) * data.sum()))
    plt.show()
```

```python
def plot_mulit_bars(data, choices) :
    plt.figure(figsize = (15, 10))
    width =  0.2

    n = data.shape[1]
    indexes = np.arange(n)
    fig, ax = plt.subplots()
    for i in range(n) :
        ax.bar(indexes + i * width, data[:,i], width, label = choices[i])
    
    for container in ax.containers:
        ax.bar_label(container)

    plt.xlabel('Choices')
    plt.ylabel('Count')
    plt.xticks(indexes + width * n / 2 - 0.1, [f'Choice {i}' for i in range(1, n + 1)])
    ax.legend(loc = 'best')
    plt.show()
```

```python
yes_no_sitamarhi_q1 = np.array([16,  8])
title_sitamarhi_q1 = sitamarhi + ' - ' + 'Were you natively living in the town, or you moved here from the village?'
print(title_sitamarhi_q1)
plot_yes_no(yes_no_sitamarhi_q1, random.sample(sns.color_palette('bright'), 2))
```

```python
yes_no_sitamarhi_q2 = np.array([12, 12])
title_sitamarhi_q2 = sitamarhi + ' - ' + 'Would you migrate in the coming years?'
print(title_sitamarhi_q2)
plot_yes_no(yes_no_sitamarhi_q2, random.sample(sns.color_palette('bright'), 2))
```

```python
orders_sitamarhi_q3 = np.array([[16,  8,  0,  0], [ 4, 12,  0,  8], [ 4,  4, 12,  4], [ 0,  0, 12, 12]])
choices = ['Economic Opportunities', 'Education', 'Infrastructure', 'Community + Society']
title_sitamarhi_q3 = sitamarhi + ' - ' + 'What would be the order of preference of following aspects - ' + ', '.join(choices) + ' ?'
print(title_sitamarhi_q3)
plot_mulit_bars(orders_sitamarhi_q3, choices)
```
